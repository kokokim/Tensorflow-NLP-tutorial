## 다중 선형 회귀 구현
* 중간고사, 기말고사, 추가점수를 어떤 공식을 통해 최종 점수를 계산한 데이터가 있음
* 예측 모델을 구현해보쟈~

### 데이터준비

```python
# 중간 고사, 기말 고사, 수행평가
import numpy as np

X = np.array([[70,85,11], [71,89,18], [50,80,20], [99,20,10], [50,10,10]])
y = np.array([73, 82 ,72, 57, 34]) # 최종 성적
```

###모델 생성~ 쌓아보자~

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers

model = Sequential()
model.add(Dense(1, input_dim=3, activation='linear'))

sgd = optimizers.SGD(learning_rate=0.0001)
model.compile(optimizer=sgd, loss='mse', metrics=['mse'])
model.fit(X, y, epochs=2000)
```

### 결과보기

```python
print(model.predict(X)) #모델이 예측값에 거의 근접하는구먼!
```

```python
#새로운 데이터로 예측해보쟝쟝쟝
X_test = np.array([[20,99,10], [40,50,20]])
print(model.predict(X_test)) #음 괜춘하군
```

## 다중 로지스틱 회귀
* 꽃받침의 길이와 꽃잎의 길이와 해당 꽃이 a인지 b인지 적혀져 있는 데이터가 있을떄, 새로 조사한 꽃받침의 길이와 꽃잎의 길이로부터 무슨 꽃인지 예측할 수 있는 모델을 만들고자 한다!!
* A 종 : 0 , B종 : 1

### 데이터 준비

```python
X = np.array([[5.1, 3.5], [4.7, 3.2], [5.2, 1.8], [7, 4.1], [5.1, 2.1]])
y = np.array([0, 0, 1, 0, 1])
```

### 모델 구현

```python
model=Sequential()
model.add(Dense(1, input_dim=2, activation='sigmoid'))
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['binary_accuracy'])

model.fit(X, y, epochs=2000)
```

### 결과 확인

```python
print(model.predict(X)) #예측을 잘 하였꾼!!! a종은 0.5 이하 b종은 0.5 이상!! 나옴
```
